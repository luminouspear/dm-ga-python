```python
#Define a class BankAccount
#A BankAccount has an id, an account_number, a name (the person who owns the account), and a balance
#The opening balance of the account is 0
class BankAccount:
    def __init__(self, id, account_number, name):
        #implement the init method
        self.name = name
        self.account_number = account_number
        self.id = id
        self.balance = 0

    def deposit(self, account, amount):
        if self.account_number == account:
            self.balance += amount
            print(f'Your new balance is ${self.balance}')
        else:
            print(f'Unable to process your request. Please check your account number.')

    def withdraw(self, account, amount):
        remainder = self.balance - amount
        if remainder >= 0 and self.account_number == account:
            self.balance = remainder
            print(f'Withdrew ${amount}, final balance is ${self.balance}.')
        else:
            print(f'Unable to withdraw ${amount} as your current balance is ${self.balance}.')

#Implement the __repr__ method to display the id, account_number, name, and balance for an account
    def __repr__(self):
        return f"\nAccount Name:\t\t{self.name}\nAccount Number:\t\t{self.account_number}\nAccount ID:\t\t{self.id}\nAccount Balance:\t${self.balance}"

account = BankAccount(1234,123142,"John Smith")

account.withdraw(123142, 40)
account.deposit(1234142, 40)
account.deposit(123142, 40)
account.withdraw(123142, 20)
account.withdraw(123142, 21)
print(account)
```

```python
#Define a class Flight that we will use to track an airline flight
# and all the bookings we have
#The Fight class has flight_number, origin, destination, capacity

class Flight:
    def __init__(self, flight_number, origin, destination, capacity):
        self.flight_number = flight_number
        self.origin = origin
        self.destination = destination
        self.capacity = capacity
        self.customers: Customer = []
        #define the init method
        # 

#define a book method that books the flight
#the book method must decrese the capacity of the flight 
#by 1 each time it is called
    def book_flight(self, number, customer):
        if self.capacity > 0 and self.flight_number == number:
            self.capacity -= 1
            self.customers.append(customer)
            print(f'Flight #{self.flight_number} booked. Current capacity on this flight is {self.capacity}.')
        else:
            print(f'Flight #{self.flight_number} is currently sold out. Let\'s sell them the ticket anyway though.')

    def get_bookings(self):
        if self.customers:
            print(f"List of Passengers on Flight #{self.flight_number}, starting in {self.origin} and ending in {self.destination}:")
            for customer in self.customers:
                print(f"Passenger Name: {customer.name}, Passenger ID: {customer.id}, Age: {customer.age}, Sex: {customer.sex}\n")
        else:
            print("No one has booked this flight yet.")
    
    def __repr__(self):
        print(f'Flight #{self.flight_number}, leaving from {self.origin} and arriving at {self.destination}.')
#define the __repr__ method that prints out a flight's number, origin, and destination

houston = Flight(1234, "Miami", "Houston", 2)


```
