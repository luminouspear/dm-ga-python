```python
#read in the credit_cards.csv file

import pandas as pd

cards = pd.read_csv('credit_cards.csv', parse_dates=['Expiry Date','Issue Date'])
```

```python
# show the top ten records in the data set

cards.head(10)
```

```python
# show a summary of the columns, their types and the non-null count of each

cards.info()
```

```python
# rename the 'card type full name' and Card Holder's Name 
# columns to 'Type' and 'Client' respectively

cards.rename(columns = { 'Card Type Full Name': 'Type', "Card Holder's Name": 'Client' }, inplace=True)
cards.info()
```

```python
# show card type, bank, client for all cases where credit 
# limit > 70 000

cards[cards['Credit Limit'] > 70000][['Type', 'Issuing Bank','Client']]
```

```python
# find all clients, billing date, card numbers for clients of Chase bank

cards[cards['Issuing Bank'] == 'Chase'][['Client','Billing Date','Card Number']]
```

```python
cards.info()
```

```python
# show all clients card numbers, credit limit for Visa with credit limit > 100 000

cards[(cards['Type'] == 'Visa') & (cards['Credit Limit'] > 100000)][['Card Number','Credit Limit']]
```

```python
# gwet all clients, issuing bank, credit limit of all
#  those that are Master Card and expiry date is after 2020.

cards[(cards['Type'] == 'Master Card') & (cards['Expiry Date'].dt.year >
                                          2020)][['Client', 'Issuing Bank', 'Credit Limit']]
```

```python
cards[(cards['Type'] == 'Master Card') & (cards['Expiry Date'].apply(lambda x: x[:].dt.year) > 2020)]
```

```python
# how many different card issuers are there?

cards['Issuing Bank'].value_counts()
```
