```python
# create dataframe w/o index

data = {'Make': ['Toyota', 'Lexus', 'BMW', 'Hyundai'], 
'Model': ['Rav 4', 'LS20', 'X3', 'Elantra'], 
'Year': [2018, 2015, 2012, 2022], 
'Color': ['white', 'black', 'blue', 'gray']}



```

```python
import pandas as pd

cars = pd.DataFrame(data)

cars
```

```python
# can you merge with a data set that isn't in the same order? 

more_data = {'Make': ['Toyota1', 'Lexus1', 'BMW1', 'Hyundai1'],
        'Year': [2018, 2015, 2012, 2022],
        'Model': ['Rav 41', 'LS201', 'X31', 'Elantra'],
        'Color': ['white1', 'black1', 'blue1', 'gray1']}

more_cars = pd.DataFrame(more_data)

all_cars = pd.concat([cars, more_cars])

```

```python
price_data = {'Price': [35000, 50000, 54000, 27000], 
'Seats': [5, 5, 5, 5]}

prices = pd.DataFrame(price_data)
```

```python
pd.concat([cars, prices], axis = 1)
```

```python
pd.concat([all_cars, prices], axis = 1)
```
