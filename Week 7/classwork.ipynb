```python
#read in the eCommerce data set

import pandas as pd

data = pd.read_csv('e_commerce.csv', parse_dates=['CC Exp Date'])

```

```python
#read the data types of the columns and show how many null values there are.

data.info()
```

```python
data.head()
```

```python
# add a column 'Provider' that displays the email provider for each customer
# add a column with data_frame['new_column'] = data_frame['old_data'] + validation

data['Provider'] = data['Email'].apply(lambda x:x.split('@')[1])

data['Provider'].head()
```

```python
def get_email_provider(email):
    return email.split('@')[1]
```

```python
data['Provider'] = data['Email'].apply(get_email_provider)

data['Provider'].head()
```

```python
# average purchase price

data['Purchase Price'].sum() / data['Purchase Price'].count()
```

```python
# max purchase price

data['Purchase Price'].max()
```

```python
# how many people have 'en' as language of choice for their browser?

data[data['Language'] == 'en'].count()
```

```python
# how many people made their purchase during the AM vs during PM?

data['AM or PM'].value_counts()
```

```python
# what are the 5 most common job titles?

data[['Job']].value_counts().head()
```

```python
# how many people have American Express as their credit card provider and madea purchase above $95

data[(data['CC Provider'] == 'American Express') & (data['Purchase Price'] > 95)].count()
```

```python
#hard: how many people have a credit card that expires in 2025?

data['Date'] = pd.to_datetime(data['CC Exp Date'], format="%m/%y")

data[data['Date'].apply(lambda x: x.year == 2025)]

#data['Date'].apply(lambda x: x.dt.year == 25)
```
