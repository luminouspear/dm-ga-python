```python
import requests

data = requests.get("https://statsapi.web.nhl.com/api/v1/teams")
```

```python
#data is a response object
#200 means OK

data
```

```python
#store team data in a JSON object

teams_data = data.json()
```

```python
teams_data
```

```python
teams_data['teams']
```

```python
for team in teams_data['teams']:
    print(team['name'])
```

```python
total = 0

for team in teams_data['teams']:
    if team['conference']['name'] == "Eastern":
        total += 1


print(total)
```

```python
division_set = set()

for team in teams_data['teams']:
    division_set.add(team['division']['name'])
print(division_set)
```

```python
firstYears = []

for team in teams_data['teams']:
    if len(firstYears) == 0:
        firstYears.append(team)
    elif firstYears[0]['firstYearOfPlay'] > team['firstYearOfPlay']:
        firstYears.pop(0)
        firstYears.append(team)

print(firstYears[0]['name'], firstYears[0]['firstYearOfPlay']) 
```

```python
years = []

for team in teams_data['teams']:
    my_team = {}
    my_team['name'] = team['name']
    my_team['firstYearOfPlay'] = team['firstYearOfPlay']
    years.append(my_team)

sorted(years, key=lambda x: x['firstYearOfPlay'])
```
